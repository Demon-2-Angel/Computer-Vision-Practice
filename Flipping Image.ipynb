```python
import cv2 
from matplotlib import pyplot as plt 
```

```python
img = cv2.imread(r"C:\Users\aniru\Machine Learning\Computer Vision\white lotus.jpg")
name = "White Lotus"
plt.title(name)
plt.imshow(img)
```

## Flipping image Horizontally

```python
horizontal_flip = cv2.flip(img,1)
name_hori = "White Lotus Horizontal Flip"
plt.title(name_hori)
plt.imshow(horizontal_flip)
```

## Flipped Image Vertically


```python
vertical_flip = cv2.flip(img, 0)
name_verti = "white Lotus Vertically Flipped"
plt.title(name_verti)
plt.imshow(vertical_flip)
```

```python
combined_flip = cv2.flip(img, -1)
name_combined = "White Lotus horizon + vertically Flipped"
plt.title(name_combined)
plt.imshow(combined_flip)
```
